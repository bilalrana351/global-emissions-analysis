```python
# Apply enrichment: Add Continent, UN Region, and Income Group columns
df['Continent'] = df['Country'].map(lambda x: COUNTRY_METADATA.get(x, {}).get('Continent', 'Unknown'))
df['UN Region'] = df['Country'].map(lambda x: COUNTRY_METADATA.get(x, {}).get('UN Region', 'Unknown'))
df['Income Group'] = df['Country'].map(lambda x: COUNTRY_METADATA.get(x, {}).get('Income Group', 'Unknown'))

print(f"Dataset enriched! New shape: {df.shape}")
print(f"\nNew columns added: {list(df.columns[-3:])}")

print(f"\nFirst few rows for countries with unique names (one row per country):")
unique_countries_df = df.drop_duplicates(subset='Country')[['Country', 'Continent', 'UN Region', 'Income Group']].head(10)
unique_countries_df
```

```python
# Verification: Check for any missing or unknown values
print("=== VERIFICATION REPORT ===\n")

# Check for 'Unknown' values in new columns
unknown_continent = df[df['Continent'] == 'Unknown']['Country'].unique()
unknown_region = df[df['UN Region'] == 'Unknown']['Country'].unique()
unknown_income = df[df['Income Group'] == 'Unknown']['Country'].unique()

print(f"Countries with Unknown Continent: {len(unknown_continent)}")
if len(unknown_continent) > 0:
    print(f"  {list(unknown_continent)}")

print(f"\nCountries with Unknown UN Region: {len(unknown_region)}")
if len(unknown_region) > 0:
    print(f"  {list(unknown_region)}")

print(f"\nCountries with Unknown Income Group: {len(unknown_income)}")
if len(unknown_income) > 0:
    print(f"  {list(unknown_income)}")

# Summary statistics
print("\n=== SUMMARY STATISTICS ===\n")
print("Continent Distribution:")
print(df['Continent'].value_counts())

print("\n\nUN Region Distribution:")
print(df['UN Region'].value_counts().head(10))

print("\n\nIncome Group Distribution:")
print(df['Income Group'].value_counts())

print("\nâœ… Enrichment completed successfully!" if len(unknown_continent) == 0 and len(unknown_region) == 0 and len(unknown_income) == 0 else "\nâš ï¸ Some countries have unknown values!")
```
